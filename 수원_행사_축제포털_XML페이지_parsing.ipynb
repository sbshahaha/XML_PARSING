```python
#라이브러리 import
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime


#요청url
url = "https://www.suwon.go.kr/web/connectUpdate/cultureListExp.do?q_currPage=1&q_rowPerPage=100"
image_base_URL ='https://www.suwon.go.kr'
current_date = datetime.now().strftime("%Y-%m-%d")
```

```python
#데이터 받아오기/확인
result = requests.get(url)
soup = BeautifulSoup(result.text,'lxml-xml')
items = soup.find_all("cultureList")

print(items)
```

```python
#parsing 할 함수정의
def parse():
    try:
        ctrSeqNo = item.find("ctrSeqNo").get_text()
        groupCd = item.find("groupCd").get_text()
        groupNm = item.find("groupNm").get_text()
        cultureNm = item.find("cultureNm").get_text()
        startDt = item.find("startDt").get_text()
        endDt = item.find("endDt").get_text()
        ctrLocation = item.find("ctrLocation").get_text()
        thumbImage = item.find("thumbImage").get_text()
        con_URL = f'https://www.suwon.go.kr/culture/ingCultureView.do?ctrSeqNo={ctrSeqNo}&listType=main&q_groupCd=&q_year=&q_month=&q_currPage=1&q_rowPerPage=12&q_ingYn=0&q_searchDatePart=&q_startDt=&q_endDt=&q_cultureNm='


        return {
              "ctrSeqNo(일련번호)":ctrSeqNo,
              "groupCd(분류코드)":groupCd,
              "groupNm(분류명)":groupNm,
              "cultureNm(행사명)":cultureNm,
              "startDt(시작일)":startDt,
              "endDt(종료일)":endDt,
              "ctrLocation(장소)":ctrLocation,
              "thumbImage(썸네일이미지)":thumbImage,
              "con_URL(연결링크)":con_URL
          }
    except AttributeError as e:
        return {
            "ctrSeqNo(일련번호)":None,
            "groupCd(분류코드)":None,
            "groupNm(분류명)":None,
            "cultureNm(행사명)":None,
            "startDt(시작일)":None,
            "endDt(종료일)":None,
            "ctrLocation(장소)":None,
            "thumbImage(썸네일이미지)":None,
            "con_URL(연결링크)":None
        }
```

```python
#데이터 받아오기
result = requests.get(url)
soup = BeautifulSoup(result.text,'lxml-xml')
items = soup.find_all("cultureList")
```

```python
#파싱한 데이터 정리
row = []
for item in items:
  if item.find("endDt").get_text() > current_date:
    row.append(parse())
  else :
    continue
```

```python
#pandas 데이터프레임에 넣기
df = pd.DataFrame(row)

#csv 파일로 저장하기
df.to_csv("수원행사축제포털.csv",mode='w',)

#csv 파일 불러오기
data = pd.read_csv("수원행사축제포털.csv",index_col=0)
df2 = pd.DataFrame(data)
```
